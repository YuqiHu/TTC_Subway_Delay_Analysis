# The first part of this jupyter notebook is cleaning the Toronto Assult Data and plot the data with Subway Station Location on Map

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Import the Assault Open Data

```python
# Read t# The first part of this jupyter notebook is cleaning the Toronto Assult Data and plot the data with Subway Station Location on Maphe Assault_Open_Data CSV into a DataFrame
criminal_data_raw = pd.read_csv('Assault_Open_Data.csv')
criminal_data_raw.head()
```

Drop the columns that are not useful

```python
#create a new data frame contains the useful columns below
#OCC_DATE
#LOCATION_TYPE
#PREMISES_TYPE
#OFFENCE
#NEImport the Assault Open DataIGHBOURHOOD_140
#LONG_WGS84
#LAT_WGS84

criminal_data_sorted = criminal_data_raw[[ 
                                        'OCC_DATE','LOCATION_TYPE','PREMISES_TYPE',
                                        'OFFENCE','NEIGHBOURHOOD_140','LONG_WGS84','LAT_WGS84']]

criminal_data_sorted.head()
```

Filter the location type to TTC subway station and TTC subway train

```python
#Sort the assult cases location that contains subway
criminal_data_subway = criminal_data_sorted[criminal_data_sorted['LOCATION_TYPE'].str.contains('Subway', case=False, na=False)]
Import the Assault Open Data
criminal_data_subway.head()
```

```python
#convert the occ_date column to datetime type
criminal_data_subway['OCC_DATE'] = pd.to_datetime(criminal_data_subway['OCC_DATE'])
```

```python
'''
make sure the occ date range match with the delay time
'''
start_date = '2014/01/01 05:00:00+00'
end_date = '2023/09/30 05:00:00+00'

# Convert the start and end dates to datetime
start_date = pd.to_datetime(start_date)
end_date = pd.to_datetime(end_date)

criminal_data_subway = criminal_data_subway[(criminal_data_subway['OCC_DATE'] >= start_date) & (criminal_data_subway['OCC_DATE'] <= end_date)]
criminal_data_subway = criminal_data_subway.set_index('OCC_DATE')
criminal_data_subway.head()
```

```python
'''
From further investigation, I think the subway delay happens at the stationn at the most especially i want to find
out the correlation with disordered patron, so I only keep the LOCATION_TYPE in Ttc Subway Train
'''
filtered_df = criminal_data_subway[criminal_data_subway['LOCATION_TYPE'] != 'Ttc Subway Train']
filtered_df.head()
```

```python
'''
The lat and lon for assualt location is different even they are in a same neighbourhood. 
In order to make it constant, I decide to use the lat and lon that first comes out for that neighbourhood
'''

first_occurrences = filtered_df.groupby('NEIGHBOURHOOD_140').first()[['LONG_WGS84', 'LAT_WGS84']]

for neighborhood in first_occurrences.index:
    longitude, latitude = first_occurrences.loc[neighborhood]
    filtered_df.loc[filtered_df['NEIGHBOURHOOD_140'] == neighborhood, 'LONG_WGS84'] = longitude
    filtered_df.loc[filtered_df['NEIGHBOURHOOD_140'] == neighborhood, 'LAT_WGS84'] = latitude

filtered_df.head(5)
```

```python
'''
Find the total assualt number for each neighbourhood subway station
'''
counts = filtered_df['NEIGHBOURHOOD_140'].value_counts()

# Converting the Series to a DataFrame
station_counts_df = counts.reset_index()
station_counts_df.columns = ['Station', 'Count']

station_counts_df.head()
```

```python
'''
assign the lon and lat to the subway station
'''
subset_df1 = filtered_df[['NEIGHBOURHOOD_140', 'LONG_WGS84', 'LAT_WGS84']]

# Rename the 'Station' column to 'NEIGHBOURHOOD_140' to match
station_counts_df.rename(columns={'Station': 'NEIGHBOURHOOD_140'}, inplace=True)

# Merge the DataFrames on the NEIGHBOURHOOD_140 column
merged_df = pd.merge(station_counts_df, subset_df1, on='NEIGHBOURHOOD_140', how='left')
merged_df = merged_df.drop_duplicates(subset='NEIGHBOURHOOD_140')
# Select the top 40 stations
merged_df = merged_df.head(40)

merged_df.head()
```

```python
#Graph the subway station location and assult count

import folium
from folium.plugins import MarkerCluster
import branca.colormap as cm

m = folium.Map(location=[43.656323, -79.380923], zoom_start=12,tiles='CartoDB positron',
               width='100%', height='200%')

# Create a color scale
color_scale = cm.linear.YlOrRd_09.scale(min(merged_df['Count']), max(merged_df['Count']))

# Add points
for _, row in merged_df.iterrows():
    folium.CircleMarker(location=[row['LAT_WGS84'], row['LONG_WGS84']],
                        radius=8,
                        color=color_scale(row['Count']),
                        fill=True,
                        fill_color=color_scale(row['Count']),
                        fill_opacity=5).add_to(m)

# Add scale to map
color_scale.add_to(m)

# Display the map
m
```

The map uses color intensity to represent the frequency of assault cases at each subway station: the darker the color, the higher the number of cases. It is evident that the top three stations with the darkest colors are located in the following neighborhoods: Church-Yonge Corridor, Annex, and Rosedale-Moore Park.

# The Second part of this jupyter notebook is plotting the disorderly patron number to the map

```python
import geopandas as gpd

# Load a shapefile for subway location
gdf = gpd.read_file('subway_stations.shp')

# only keep the columns that I need
selected_df = gdf[['STATION', 'geometry']]
selected_df.head()
```

```python
# Make the station name to uppercase, so I can apply groupby at latter stage

selected_df['STATION'] = selected_df['STATION'].str.upper()
selected_df.head()
```

```python
# Read the Delay CSV into a DataFrame
delay = pd.read_csv('TTC_Real_Delay_2014_to_2023.csv')

# Filter the delay reason with Disorderly Patron
filtered_code = delay[delay['Code'] == 'SUDP']
filtered_code.head()
```

```python
# Find the Disorderly Patron number at each station
delay_stn = filtered_code['Station'].value_counts()

# Converting the Series to a DataFrame
delay_stn = delay_stn.reset_index()
delay_stn.columns = ['Station', 'Count']

delay_stn.head()
```

```python
# Clean the delay_stn Station name by cutting off the 'STATION' from it
delay_stn['Station'] = delay_stn['Station'].str.replace(" STATION", "", regex=False)

# Mrger the delay_stn df and selected_df together, so we would have Geometry information on the table
merged_df_delay = pd.merge(delay_stn, selected_df, left_on='Station', right_on='STATION', how='left')

# There are also some rows that can not merge together, so I need to do it manually 
merged_df_delay = merged_df_delay.head(65)
merged_df_delay.at[0, 'geometry'] = selected_df.at[17, 'geometry']
merged_df_delay.at[2, 'geometry'] = selected_df.at[30, 'geometry']
merged_df_delay.at[3, 'geometry'] = selected_df.at[17, 'geometry']
merged_df_delay.at[5, 'geometry'] = selected_df.at[15, 'geometry']
merged_df_delay.at[7, 'geometry'] = selected_df.at[15, 'geometry']
merged_df_delay.at[16, 'geometry'] = selected_df.at[44, 'geometry']
merged_df_delay.at[19, 'geometry'] = selected_df.at[14, 'geometry']
merged_df_delay.at[26, 'geometry'] = selected_df.at[49, 'geometry']
merged_df_delay.at[30, 'geometry'] = selected_df.at[14, 'geometry']
merged_df_delay.at[35, 'geometry'] = selected_df.at[49, 'geometry']
merged_df_delay.at[50, 'geometry'] = selected_df.at[50, 'geometry']
merged_df_delay = merged_df_delay.dropna(subset=['geometry'])

merged_df_delay.head()
```

```python
#Find out the counnt if they are have the same Geometry
delay_df = merged_df_delay[['geometry','Count']]
delay_df.head()
```

```python
'''
Convert the geometry into Latitude and Longitude
'''
import pandas as pd
from pyproj import Proj, Transformer
from shapely.geometry import Point

utm_zone = 17
is_south = False

# Create a Transformer object for UTM to WGS84 coordinate transformation
transformer = Transformer.from_proj(
    Proj(proj='utm', zone=utm_zone, ellps='WGS84', south=is_south),
    Proj(proj='latlong', datum='WGS84')
)

# Function to convert UTM to Lat/Lon
def utm_to_latlon(point):
    if point and isinstance(point, Point):
        lon, lat = transformer.transform(point.x, point.y)
        return lat, lon
    else:
        return None, None

# Apply the function to the 'geometry' column
delay_df[['Latitude', 'Longitude']] = delay_df['geometry'].apply(utm_to_latlon).apply(pd.Series)

delay_df.head()
```

```python
import folium
import branca.colormap as cm

# Create a base map
y = folium.Map(location=[43.656323, -79.380923], zoom_start=12, tiles='CartoDB positron')

# Create a color scale
min_count = delay_df['Count'].min()
max_count = delay_df['Count'].max()
color_scale = cm.linear.Blues_09.scale(min_count, max_count)

# Add points
for _, row in delay_df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=8,
        color=color_scale(row['Count']),
        fill=True,
        fill_color=color_scale(row['Count']),
        fill_opacity=0.7
    ).add_to(y)

# Add color scale to map
color_scale.add_to(y)

# Display the map
y
```

This map displays the incidence of disorderly patron delays at TTC subway stations. Similar to the assault case data, we can examine the correlation between the number of disorderly patron delays and assault cases. This comparison allows us to understand the interrelation between these two metrics. For instance, the Bloor-Yonge station ranks first in both assault cases and disorderly patron incidents.
