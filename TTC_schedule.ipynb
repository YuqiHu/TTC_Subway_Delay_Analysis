# Web Scrap TTC Line 1 and Line 2 Schedule at St.George and Bloor-Yonge Stations

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

## 1. Create a function to scrap schedule from the website

```python
def get_TTC_scheduled_time(url):
    # Set up the driver
    driver = webdriver.Edge()

    # Get the webpage
    driver.get(url)

    # Create an empty list
    time = []

    # Run through all elements dynamically
    i = 1
    while True:
        try:
            # Set the element for extracting. Extract the information by the XPATH
            element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, f"/html/body/mv-root/div/div[1]/main/mv-line-router/mv-stop-arrivals/div[2]/div[{i}]/span"))
            )
            # Extract the text content of the element and append the list
            time.append(element.text)
            i += 1
        except Exception as e:
            # Break the loop if no more elements are found
            break

    # Close the website
    driver.quit()

    return time
```

## 3. Find the percentage of the matching schedule

### St.George Station

```python
# Calculate the number of time slots are the same between line 1 and line 2 at St.George Station
st_george_common_time = len(set(st_george_line1_to_Finch_time).intersection(set(st_george_line2_to_Kennedy_time))) + len(set(st_george_line1_to_Finch_time).intersection(set(st_george_line2_to_Kipling_time))) + len(set(st_george_line1_to_Vaughan_time).intersection(set(st_george_line2_to_Kennedy_time))) + len(set(st_george_line1_to_Vaughan_time).intersection(set(st_george_line2_to_Kipling_time)))
print("There are {} time slots are the same for line 1 and line 2 in schedule at St.George Station, which is {}%.".format(st_george_common_time, round(st_george_common_time/(len(st_george_line1_to_Finch_time) + len(st_george_line1_to_Vaughan_time) + len(st_george_line2_to_Kennedy_time) + len(st_george_line2_to_Kipling_time))*100, 2)))
```

### Bloor-Yonge Station

```python
# Calculate the number of time slots are the same between line 1 and line 2 at Bloor Yonge Station
bloor_common_time = len(set(bloor_line1_to_Finch_time).intersection(set(bloor_line2_to_Kennedy_time))) + len(set(bloor_line1_to_Finch_time).intersection(set(bloor_line2_to_Kipling_time))) + len(set(bloor_line1_to_Vaughan_time).intersection(set(bloor_line2_to_Kennedy_time))) + len(set(bloor_line1_to_Vaughan_time).intersection(set(bloor_line2_to_Kipling_time)))
print("There are {} time slots are the same for line 1 and line 2 in schedule at Bloor-Yonge station, which is {}%.".format(bloor_common_time, round(bloor_common_time/(len(bloor_line1_to_Finch_time) + len(bloor_line1_to_Vaughan_time) + len(bloor_line2_to_Kennedy_time) + len(bloor_line2_to_Kipling_time))*100, 2)))
```
